```python
import cv2
import matplotlib.pyplot as plt
import imageio
from PIL import Image
```

```python
img_cv2 = cv2.imread('image.JPG')
img_pil = Image.open('image.JPG')
img_matplotlib = plt.imread('image.JPG')
img_imageio = imageio.imread('image.JPG')
```

```python
plt.imshow(cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB))
plt.title('OpenCV Image')
plt.axis('off')
plt.show()

img_pil.show()

plt.imshow(img_matplotlib)
plt.title('Matplotlib Image')
plt.axis('off')
plt.show()

plt.imshow(img_imageio)
plt.title('ImageIO Image')
plt.axis('off')
plt.show()
```

```python
img = Image.open('image.JPG')
img.save('image.png')
img.save('image.tif')
```
