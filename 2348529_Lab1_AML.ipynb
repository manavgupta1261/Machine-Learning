```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/content/Program_1_Dataset - Program_1_Dataset.csv')
```

```python
predictor = df['Avg_rooms'].values.reshape(-1, 1)
target = df['MValue'].values
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Avg_rooms', y='MValue', data=df)
plt.title('Scatter Plot: Avg_rooms vs MValue')
plt.xlabel('Avg_rooms')
plt.ylabel('MValue')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictor, target, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
coef = model.coef_[0]
intercept = model.intercept_
print(f'Coefficient: {coef:.2f}')
print(f'Intercept: {intercept:.2f}')
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
```

```python
print(f'Mean Squared Error (Train): {mse_train:.2f}')
print(f'Mean Squared Error (Test): {mse_test:.2f}')
print(f'R-squared (Train): {r2_train:.2f}')
print(f'R-squared (Test): {r2_test:.2f}')
```
